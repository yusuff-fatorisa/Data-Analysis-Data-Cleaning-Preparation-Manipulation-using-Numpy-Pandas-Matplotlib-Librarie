```python
regions94.index
```

```python
new_index = list(regions94.index)
new_index
regions_series.index = new_index
regions_series.index.name = 'who_region'
regions_series.columns.name = "Years"
```

```python
regions_series
```

```python
regs_series = regions_series.T
regs_series
```
